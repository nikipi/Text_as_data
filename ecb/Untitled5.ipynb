```python

import pandas as pd
import numpy as np
import ssl


url = '/Users/ypi/Desktop/ecb/pred.csv'

pred= pd.read_csv(url)
pred.tail()
```

```python
a=[]
for i in range(1,583):
    strg='text{}\\.'.format(i)
    a.append(strg)

```

```python
subdf=pred[pred['text_sens'].str.contains("text1\\.")]
print(len(subdf))
print(len(subdf[subdf['fit']>=0]))
print(len(subdf[subdf['fit']<=0]))
```

```python
subdf=pred[pred['text_sens'].str.contains("text3\\.")]
print(len(subdf))
print(len(subdf[subdf['fit']>=0]))
print(len(subdf[subdf['fit']<0]))
```

```python
num_sens=[]
num_hawk_sens=[]
num_dove_sens=[]
for item in a:
    subdf=pred[pred['text_sens'].str.contains(item)]
    num_sens.append(len(subdf))
    num_hawk_sens.append(len(subdf[subdf['fit']>=0]))
    num_dove_sens.append(len(subdf[subdf['fit']<0]))
```

```python
num_hawk_sens
```

```python

url = '/Users/ypi/Desktop/ecb/ecb_whole_new.csv'

ecb= pd.read_csv(url)
ecb.tail()
```

```python
ecb['num_sens']=num_sens
ecb['num_hawk_sens']=num_hawk_sens
ecb['num_dove_sens']=num_dove_sens
```

```python
ecb['lss_sens']=np.where(ecb['num_hawk_sens']>=ecb['num_dove_sens'],'Hawk','Dove')
```
